```python
import numpy as np
import matplotlib.pyplot as plt

n=100000
xfunction=np.array([])

for i in range(n):
    x1=np.random.uniform(-1,1)
    x2=np.random.uniform(0,3)

    while x2 >=3*x1**2:
        x1=np.random.uniform(-1,1)
        x2=np.random.uniform(0,3)
    xfunction = np.append(xfunction, x1)

print(xfunction)
X=np.arange(-1, 1, 0.01)
fx=1.5*X**2
plt.plot(X,fx)
plt.hist(xfunction,bins=50,density=True)
plt.show()




```
