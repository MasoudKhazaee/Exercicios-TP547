```python
import numpy as np
import matplotlib.pyplot as plt
N = 10000
p = 0.4
count = 0
x_gen = np.array([])


geometric = 0
for n in range(N):
  u = np.random.uniform(0,1)
  if u >= p:
    geometric = geometric + 1
  else:
    x_gen = np.append(x_gen,geometric)
    geometric = 0

for k in x_gen:
  if k == 5:
    count = count + 1

print(x_gen)

x = np.arange(np.min(x_gen),np.max(x_gen),1)
f_x = p*(1-p)**(x-1)

p_6 = count/len(x_gen)
p_6t = p*(1-p)**(6-1)

print("Probabilidade na sexta tentativa",p_6)
print("A probabilidade teórica correspondente é:",p_6t)



plt.plot(x, f_x, color = 'red')
plt.hist(x_gen, bins = 50,density = 'True', color = 'green', label = 'Geometric distribution')
plt.legend()
plt.show()


```
