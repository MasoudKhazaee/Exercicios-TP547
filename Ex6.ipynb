```python
import numpy as np
import matplotlib.pyplot as plt


# Redefinir a função gerar_variavel_geometrica
def gerar_variavel_geometrica(p):
    x = 1 # Inicia contagem do número de tentativas
    while np.random.random() >= p: # Gera um número aleatório entre 0 e 1; continua se não for sucesso
        x += 1
    return x

# Definir a probabilidade de sucesso (retirar uma bola preta)
p = 0.4

# Calcular a probabilidade para x = 6
f_6 = p * (1 - p) ** (6 - 1)
print('f_6=',f_6)

# Gerar novamente uma amostra de 10000 variáveis aleatórias geométricas com p = 0.4
amostra = [gerar_variavel_geometrica(0.4) for _ in range(10000)]

# Plotar o histograma da amostra
plt.hist(amostra, bins=max(amostra), edgecolor='k', alpha=0.7)
plt.xlabel('Número de Tentativas até o Primeiro Sucesso')
plt.ylabel('Frequência')
plt.xlim(1, max(amostra)) # Ajustar os limites do eixo X para melhor visualização
plt.grid(axis='y', alpha=0.75)
plt.show()


```
