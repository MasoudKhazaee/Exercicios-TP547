```python

import numpy as np
import matplotlib.pyplot as plt


N = 100000
U = np.random.uniform(0,1,N)
count = 0
lambda1 = 1/28

x_gen = -np.log(U)/(lambda1)
x = np.arange(0,np.max(x_gen),0.01)
x_t = lambda1*np.exp(-lambda1*x)

for k in x_gen:
  if k <= 4:
    count = count + 1

p = count/N

print("Probabilidade de se ter quatro chamadas ou menos:")
print(p)

plt.plot(x,x_t,color = 'red')
plt.hist(x_gen, bins = 40, density = 'True', color = 'green', label = 'Exponential distribution')
plt.legend()
plt.show()
```
